```python
##도미 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
## matplotlib-데이터 시각화 상자, pyplot-그래프 그리는 기능
import matplotlib.pyplot as plt

## scatter-산점도, xlabel-x축, ylabel-y축, show-그래프 출력
plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
##빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

##도미,빙어 그래프 그리기
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
##머신 러닝 프로그램
##생선 데이터를 하나로 합치는 단계
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
##length 리스트와 weight 리스트를 (길이, 무게) 형태로 하나씩 묶어서 튜플로 만들어줘!- 입력데이터=x 준비
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
print(fish_data)
```

```python
##도미데이터 35개, 빙어데이터 14개 전체 묶음 만듬, 정답데이터=y 준비
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
##scikit-learn 라이브러리에서 K-최근접 이웃(KNN) 모델 가져옴. 
##KNN은 가장 가까운 이웃들을 보고 그 물고기가 도미인지 빙어인지 판단해주는 방식
from sklearn.neighbors import KNeighborsClassifier
```

```python
##KNN 모델(분류기) 객체 생성
kn = KNeighborsClassifier()
```

```python
##모델 학습! (입력데이터, 정답), fit은 모델에게 데이터-생선 관계 알려주는 단계
kn.fit(fish_data, fish_target)
```

```python
##모델의 정확도 계산! fish_data를 넣었을 때 target 값을 얼마나 잘 맞추는지. (정확도는 0~1 사이로 나옴)
kn.score(fish_data, fish_target)
```

```python
##이 30, 무게 600짜리 물고기를 모델에게 보여주고 도미(1)인지 빙어(0)인지 예측하는 코드
kn.predict([[30,600]])
```

```python
## KNN이 학습할 때 기억해 둔 '훈련 데이터 X' 전체 출력
print(kn._fit_X)
```

```python
##모델이 저장해둔 정답 목록 출력
print(kn._y)
```

```python
##이웃 개수를 49개로 설정한 새로운 KNN 모델 생성
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
##새 모델로 다시 학습시킨 후 정확도 계산, 1이 안나옴
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
##정확도가 전체 49개 중 도미 확률임
print(35/49)
```

```python
## kn=기본 값. 주변 5개 데이터만 보고 판단
## kn49= 전체 데이터 49개를 이웃으로 사용. 49개 모두 비교
## kn49가 정확도가 더 좋을 것 같지만 전체 데이터에서는 도미가 대부분이므로 도미로 인식하는 확률이 많음
```
